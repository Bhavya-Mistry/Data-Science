```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("p1.xlsx")
df.head()
```

```python
le = LabelEncoder()
```

```python
df["Airline"].unique()
```

```python
df["Airline_encoded"] = le.fit_transform(df["Airline"])
df.drop("Airline",axis=1, inplace=True)
```

```python
df["Source_encoded"] = le.fit_transform(df["Source"])
df.drop("Source", axis=1, inplace=True)
```

```python
df['Additional_Info_encoded'] = le.fit_transform(df["Additional_Info"])
df.drop("Additional_Info", inplace = True, axis=1)
```

```python
df["Destination_encoded"] = le.fit_transform(df["Destination"])
df.drop("Destination", axis = 1, inplace=True)
```

```python
df.isnull().sum()
```

```python
print(df["Total_Stops"].unique())

df.dropna(inplace=True)

df.isnull().sum()
```

```python
df["Total_Stops"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['non-stop', '2 stops', '1 stop', '3 stops', '4 stops']])

df["Total_Stops_encoded"] = oe.fit_transform(df[["Total_Stops"]])
df.drop("Total_Stops", axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Date_of_Journey"].unique()
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], format='%d/%m/%Y')

df.head()

df['Year'] = df['Date_of_Journey'].dt.year
df['Month'] = df['Date_of_Journey'].dt.month
df['Day'] = df['Date_of_Journey'].dt.day
df['DayOfWeek'] = df['Date_of_Journey'].dt.day_name()

df.drop("Date_of_Journey",axis=1,inplace=True)

df.head()
```

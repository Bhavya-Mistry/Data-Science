```python
import pandas as pd
```

```python
# 1) Read The dataset
df = pd.read_csv("cars_sampled.csv")
```

```python
# 2) Make a deep copy of the dataset
deep_copied_df = df.copy(deep=True)
```

```python
# 3) Get structure of the data and mention datatype of each column
df.info()
```

```python
# 4) Get summary of the data
df.describe()
```

```python
# 5) Round all the decimal values till 3 places after decimal only
df = df.round(3)
```

```python
# 6) Display all the columns of this dataset
df.columns
```

```python
# 7) Drop unwanted columns [name, dateCrawled, dateCreated, PostalCode, lastSeen]
df.drop(columns=['name','dateCrawled','dateCreated','postalCode','lastSeen'], inplace=True)
```

```python
# 8) Remove duplicate records keep first occurrence of this
df = df.drop_duplicates()
```

```python
# 9) Sort YearofRegistration column based on index value
sorted_years = df["yearOfRegistration"].sort_index()
```

```python
# 10) Find the frequency of car based on price i.e. at the price of 1$ how many cars have been sold
price_frequency = df["price"].value_counts().sort_index()
```

```python
# 11) Find the frequency of car based on the PowerPS.
powerps_frequency = df["powerPS"].value_counts().sort_index()
```

```python
# 12) Make sure the working range of data should be
# a) YearOfRegistration < 2018 and YearOfRegistration>1950
# b) Price >= 100 and Price < =150000
# c) powerPS >=10 and PowerPS <= 500
filtered_df = df[
    (df["yearOfRegistration"] > 1950) & (df["yearOfRegistration"] < 2018) &
    (df["price"] >= 100) & (df["price"] <= 150000) &
    (df["powerPS"] >= 10) & (df["powerPS"] <= 500)
]
filtered_df.head() # Display first few rows of filtered data
```
